```python
df = pd.read_csv('netflix_titles.csv')

df
```

```python
# Анализ и обработка пропущенных значений
print(df.isnull().sum())
```

```python
# Обработка пропусков
# Заполнение пропусков в столбцах 'director', 'cast', 'country' значением 'Unknown'
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')

# Удаление строк с пропущенными значениями в столбце 'date_added', т.к. их немного (10)
df.dropna(subset=['date_added'], inplace=True)

# Заполнение пропущенных значений в столбце 'rating' и 'duration' наиболее частым значением
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
df['duration'] = df['duration'].fillna(df['duration'].mode()[0])

# Удалим столбец description, так как он, вероятнее всего, не понадобится для последующей работы
df = df.drop('description', axis=1)

print(df.isnull().sum()) #Проверка, что пропусков больше нет.
```

```python
# Удаление дубликатов
num_duplicates = df.duplicated().sum()
print(f"\nКоличество дубликатов: {num_duplicates}")
```

```python
# Преобразование типов данных
#Преобразуем 'date_added' в формат даты, обрабатывая смешанные форматы
df['date_added'] = pd.to_datetime(df['date_added'], format='mixed')

df['day_added'] = df['date_added'].dt.day.astype('Int64')
df['month_added'] = df['date_added'].dt.month.astype('Int64')
df['year_added'] = df['date_added'].dt.year.astype('Int64')

# df = df.drop('date_added', axis=1)

print("\nТипы данных после обработки:")
print(df.dtypes)
```

```python

df[['duration_int', 'duration_type']] = df['duration'].str.split(' ', expand=True)
df['duration_int'] = pd.to_numeric(df['duration_int'], errors='coerce')
```

```python
def split_listed_in(df):

  all_genres = set()
  for genres in df['listed_in'].str.split(', '):
    if isinstance(genres, list):
        all_genres.update(genres)
    else:
        print(f"Unexpected value type in 'listed_in': {type(genres)}, value: {genres}")

  for genre in all_genres:
    if isinstance(genre, str):
        df[genre] = df['listed_in'].apply(lambda x: 1 if isinstance(x, str) and genre in x else 0)
    else:
        print(f"Skipping invalid genre: {genre}")

  df = df.drop('listed_in', axis=1)

  return df

df = split_listed_in(df)
```
